```python
import os
from zipfile import ZipFile
import pandas as pd
data_dir = '/mnt/sdb/rawdata/wto'
#import plotly.express as px
os.chdir("/home/anderson/Dropbox/Analytics_Project")
```

```python
def extract_readin(zip_i):
    with ZipFile(zip_i, 'r') as zip:
            for file_i in zip.namelist():
                zip.extract( file_i )
                try:
                    csv = pd.read_csv(os.path.join(data_dir, file_i), na_values = '-', encoding = "ISO-8859-1", low_memory=False)
                except pd.errors.ParserError:
                    repair_procedure(file_i)
                    csv = pd.read_csv(os.path.join(data_dir,'repair_' + file_i), na_values = '-', encoding = "ISO-8859-1", low_memory=False)
                    os.remove(os.path.join(data_dir,file_i))
    return csv

dict_of_fixes = {} #key, value = fix, problem
dict_of_fixes['"Discs, tapes, solidstate nonvolatile storage devices, \'smart cards\' and other media for the recording of sound or of other phenomena, whether or not recorded, including matrices and masters for the pr"'] = '"Discs, tapes, solidstate nonvolatile storage devices, "smart cards" and other media for the recording of sound or of other phenomena, whether or not recorded, including matrices and masters for the pr"'
```

```python
def repair_procedure(file_i):
    print('repairing',file_i)
    with open(file_i, 'r', encoding = "ISO-8859-1") as f_in, open('repair_' + file_i, 'w') as f_out:
        # Write header unchanged
        header = f_in.readline()
        print(header)
        f_out.write(header)
        # Transform the rest of the lines
        for line in f_in:
            line_out = line
            for key, search in dict_of_fixes.items():
                if search in line:
                    line_out = line_out.replace(search, key)
            f_out.write(line_out)
```

```python
def load_data(df,table_name_in, modify = 'append'):
    import csv
    from io import StringIO
    from sqlalchemy import create_engine
    from datetime import datetime
    time_now = datetime.now()
    def psql_insert_copy(table, conn, keys, data_iter):
        # gets a DBAPI connection that can provide a cursor
        dbapi_conn = conn.connection
        with dbapi_conn.cursor() as cur:
            s_buf = StringIO()
            writer = csv.writer(s_buf)
            writer.writerows(data_iter)
            print('df written to string buffer')
            s_buf.seek(0)
    
            columns = ', '.join('"{}"'.format(k) for k in keys)
            if table.schema:
                table_name = '{}.{}'.format(table.schema, table.name)
            else:
                table_name = table.name
            print('sql statement prepared')
            sql = 'COPY {} ({}) FROM STDIN WITH CSV'.format(
                table_name, columns)
            print('starting sql copy!',str(time_now)[:-7])
            cur.copy_expert(sql=sql, file=s_buf)
            s_buf.close()
        return True
    
    engine = create_engine('postgresql+psycopg2://wto_writer:wto_db_create@127.0.0.1:5433/wto')
    
    df.to_sql(table_name_in, engine, method = psql_insert_copy, if_exists = modify, index = False)
    engine.dispose()
    time_now_2 = datetime.now()
    length = time_now_2 - time_now
    print("SQL upload complete - length: " + str(length)[:-7] + " time: " + str(time_now_2)[:-7])
```
